```python
import numpy
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import mnist
```

```python
# Loads the dataset from MNIST into X_train, y_train, X_test, and y_test
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# We can see the shapes of the datasets here, its in the form (num of images, pixels by pixels)
X_train.shape
X_test.shape
```

```python
# Show the images as an example
plt.subplot(221)
plt.imshow(X_train[0], cmap=plt.get_cmap('gray'))
plt.subplot(222)
plt.imshow(X_train[1], cmap=plt.get_cmap('gray'))
plt.subplot(223)
plt.imshow(X_train[2], cmap=plt.get_cmap('gray'))
plt.subplot(224)
plt.imshow(X_train[3], cmap=plt.get_cmap('gray'))
# show the plot
plt.show()
```

```python
# Calculates the total number of pixels that is being used by the model
num_pixels = X_train.shape[1] * X_train.shape[2]
print(num_pixels)

# Reshapes the data more cleanly to be inputted into model
X_train = X_train.reshape(X_train.shape[0], num_pixels).astype('float32')
X_test = X_test.reshape(X_test.shape[0], num_pixels).astype('float32')
X_train = X_train / 255
X_test = X_test / 255

# We can check the shape of it this time, which in condensed into pixels area
X_train.shape
```

```python
# Adds noise to the images now in order to train model with noisy images
noise_factor = 0.2
x_train_noisy = X_train + noise_factor * numpy.random.normal(loc=0.0, scale=1.0, size=X_train.shape)
x_test_noisy = X_test + noise_factor * numpy.random.normal(loc=0.0, scale=1.0, size=X_test.shape)
x_train_noisy = numpy.clip(x_train_noisy, 0., 1.)
x_test_noisy = numpy.clip(x_test_noisy, 0., 1.)
```

```python
# Trains the model with a Sequential Autoencoder
model = Sequential()
model.add(Dense(500, input_dim=num_pixels, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(300, activation='relu'))
model.add(Dense(500, activation='relu'))
model.add(Dense(784, activation='sigmoid'))
```

```python
# Compiles and runs the model
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train_noisy, X_train, validation_data=(x_test_noisy, X_test), epochs=2, batch_size=200)
```

```python
# Has the prediction of the noisy data and feeds that into the model
pred = model.predict(x_test_noisy)
pred.shape
```

```python
# Plots all images by resizing as needed
X_test = numpy.reshape(X_test, (10000, 28, 28)) * 255
pred = numpy.reshape(pred, (10000, 28, 28)) * 255
x_test_noisy = numpy.reshape(x_test_noisy, (-1, 28, 28)) * 255
plt.figure(figsize=(20, 4))

print("Test Images")
for i in range(10, 20, 1):
    plt.subplot(2, 10, i+1)
    plt.imshow(X_test[i,:,:], cmap='gray')
    curr_lbl = y_test[i]
    plt.title("(Label: " + str(curr_lbl) + ")")
plt.show()

plt.figure(figsize=(20, 4))
print("Test Images with Noise")
for i in range(10, 20, 1):
    plt.subplot(2, 10, i+1)
    plt.imshow(x_test_noisy[i,:,:], cmap='gray')
plt.show()

plt.figure(figsize=(20, 4))
print("Reconstruction of Noisy Test Images")
for i in range(10, 20, 1):
    plt.subplot(2, 10, i+1)
    plt.imshow(pred[i,:,:], cmap='gray')
plt.show()
```
